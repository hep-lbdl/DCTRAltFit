## Process Data

```python
data_dir = '/data1/users/asuresh/DCTRFitting/TopQuarkFitting/'
```

```python
thetas = np.linspace(170, 180, 21)

cor_mjjj_part = np.empty((0, len(thetas)))
lvals_mjjj_part = np.empty((0, len(thetas)))
vlvals_mjjj_part = np.empty((0, len(thetas)))
AUC_mjjj_part = np.empty((0, len(thetas)))

cor_mjjj_det = np.empty((0, len(thetas)))
lvals_mjjj_det= np.empty((0, len(thetas)))
vlvals_mjjj_det = np.empty((0, len(thetas)))
AUC_mjjj_det = np.empty((0, len(thetas)))

cor_4Obs_part = np.empty((0, len(thetas)))
lvals_4Obs_part = np.empty((0, len(thetas)))
vlvals_4Obs_part = np.empty((0, len(thetas)))
AUC_4Obs_part = np.empty((0, len(thetas)))

cor_4Obs_det = np.empty((0, len(thetas)))
lvals_4Obs_det= np.empty((0, len(thetas)))
vlvals_4Obs_det = np.empty((0, len(thetas)))
AUC_4Obs_det = np.empty((0, len(thetas)))

cor_6Obs_part = np.empty((0, len(thetas)))
lvals_6Obs_part = np.empty((0, len(thetas)))
vlvals_6Obs_part = np.empty((0, len(thetas)))
AUC_6Obs_part = np.empty((0, len(thetas)))

cor_6Obs_det = np.empty((0, len(thetas)))
lvals_6Obs_det= np.empty((0, len(thetas)))
vlvals_6Obs_det = np.empty((0, len(thetas)))
AUC_6Obs_det = np.empty((0, len(thetas)))
```

```python
for i in range(120):
    file_label = data_dir +"part_mjjjTrials/DCTR_1D_mT-part_mjjj-Trial{}.npz".format(i)
    file = np.load(file_label)
    
    cor_mjjj_part = np.append(cor_mjjj_part, np.expand_dims(file['cor'], axis = 0), axis = 0)
    lvals_mjjj_part = np.append(lvals_mjjj_part, np.expand_dims(file['lvals'], axis = 0), axis = 0)
    vlvals_mjjj_part = np.append(vlvals_mjjj_part, np.expand_dims(file['vlvals'], axis = 0), axis = 0)
    AUC_mjjj_part = np.append(AUC_mjjj_part, np.expand_dims(file['AUC'], axis = 0), axis = 0)
    
for i in range(120):
    file_label = data_dir +"mjjjTrials/DCTR_1D_mT-mjjj-Trial{}.npz".format(i)
    file = np.load(file_label)
    
    cor_mjjj_det = np.append(cor_mjjj_det, np.expand_dims(file['cor'], axis = 0), axis = 0)
    lvals_mjjj_det = np.append(lvals_mjjj_det, np.expand_dims(file['lvals'], axis = 0), axis = 0)
    vlvals_mjjj_det = np.append(vlvals_mjjj_det, np.expand_dims(file['vlvals'], axis = 0), axis = 0)
    AUC_mjjj_det = np.append(AUC_mjjj_det, np.expand_dims(file['AUC'], axis = 0), axis = 0)

for i in range(120):
    file_label = data_dir +"part_4ObsTrials/DCTR_1D_mT-part_4Obs-Trial{}.npz".format(i)
    file = np.load(file_label)
    
    cor_4Obs_part = np.append(cor_4Obs_part, np.expand_dims(file['cor'], axis = 0), axis = 0)
    lvals_4Obs_part = np.append(lvals_4Obs_part, np.expand_dims(file['lvals'], axis = 0), axis = 0)
    vlvals_4Obs_part = np.append(vlvals_4Obs_part, np.expand_dims(file['vlvals'], axis = 0), axis = 0)
    AUC_4Obs_part = np.append(AUC_4Obs_part, np.expand_dims(file['AUC'], axis = 0), axis = 0)

for i in range(120):
    file_label = data_dir +"4ObsTrials/DCTR_1D_mT-4Obs-Trial{}.npz".format(i)
    file = np.load(file_label)
    
    cor_4Obs_det = np.append(cor_4Obs_det, np.expand_dims(file['cor'], axis = 0), axis = 0)
    lvals_4Obs_det = np.append(lvals_4Obs_det, np.expand_dims(file['lvals'], axis = 0), axis = 0)
    vlvals_4Obs_det = np.append(vlvals_4Obs_det, np.expand_dims(file['vlvals'], axis = 0), axis = 0)
    AUC_4Obs_det = np.append(AUC_4Obs_det, np.expand_dims(file['AUC'], axis = 0), axis = 0)
    
for i in range(120):
    file_label = data_dir +"part_6ObsTrials/DCTR_1D_mT-part_6Obs-Trial{}.npz".format(i)
    file = np.load(file_label)
    
    cor_6Obs_part = np.append(cor_6Obs_part, np.expand_dims(file['cor'], axis = 0), axis = 0)
    lvals_6Obs_part = np.append(lvals_6Obs_part, np.expand_dims(file['lvals'], axis = 0), axis = 0)
    vlvals_6Obs_part = np.append(vlvals_6Obs_part, np.expand_dims(file['vlvals'], axis = 0), axis = 0)
    AUC_6Obs_part = np.append(AUC_6Obs_part, np.expand_dims(file['AUC'], axis = 0), axis = 0)

for i in range(120):
    file_label = data_dir +"6ObsTrials/DCTR_1D_mT-6Obs-Trial{}.npz".format(i)
    file = np.load(file_label)
    
    cor_6Obs_det = np.append(cor_6Obs_det, np.expand_dims(file['cor'], axis = 0), axis = 0)
    lvals_6Obs_det = np.append(lvals_6Obs_det, np.expand_dims(file['lvals'], axis = 0), axis = 0)
    vlvals_6Obs_det = np.append(vlvals_6Obs_det, np.expand_dims(file['vlvals'], axis = 0), axis = 0)
    AUC_6Obs_det = np.append(AUC_6Obs_det, np.expand_dims(file['AUC'], axis = 0), axis = 0)
```

```python
lvals_mjjj_part_mean = np.mean(lvals_mjjj_part, axis = 0)
vlvals_mjjj_part_mean = np.mean(vlvals_mjjj_part, axis = 0)
cor_mjjj_part_mean = np.mean(cor_mjjj_part, axis = 0)
AUC_mjjj_part_mean = np.mean(AUC_mjjj_part, axis = 0)

lvals_mjjj_det_mean = np.mean(lvals_mjjj_det, axis = 0)
vlvals_mjjj_det_mean = np.mean(vlvals_mjjj_det, axis = 0)
cor_mjjj_det_mean = np.mean(cor_mjjj_det, axis = 0)
AUC_mjjj_det_mean = np.mean(AUC_mjjj_det, axis = 0)

lvals_4Obs_part_mean = np.mean(lvals_4Obs_part, axis = 0)
vlvals_4Obs_part_mean = np.mean(vlvals_4Obs_part, axis = 0)
cor_4Obs_part_mean = np.mean(cor_4Obs_part, axis = 0)
AUC_4Obs_part_mean = np.mean(AUC_4Obs_part, axis = 0)

lvals_4Obs_det_mean = np.mean(lvals_4Obs_det, axis = 0)
vlvals_4Obs_det_mean = np.mean(vlvals_4Obs_det, axis = 0)
cor_4Obs_det_mean = np.mean(cor_4Obs_det, axis = 0)
AUC_4Obs_det_mean = np.mean(AUC_4Obs_det, axis = 0)

lvals_6Obs_part_mean = np.mean(lvals_6Obs_part, axis = 0)
vlvals_6Obs_part_mean = np.mean(vlvals_6Obs_part, axis = 0)
cor_6Obs_part_mean = np.mean(cor_6Obs_part, axis = 0)
AUC_6Obs_part_mean = np.mean(AUC_6Obs_part, axis = 0)

lvals_6Obs_det_mean = np.mean(lvals_6Obs_det, axis = 0)
vlvals_6Obs_det_mean = np.mean(vlvals_6Obs_det, axis = 0)
cor_6Obs_det_mean = np.mean(cor_6Obs_det, axis = 0)
AUC_6Obs_det_mean = np.mean(AUC_6Obs_det, axis = 0)
```

```python
plt.title("Particle Level AUC (120 Run Average)")

plt.plot(thetas, AUC_mjjj_part_mean, label='$m_{jjj}$')
plt.plot(thetas, AUC_4Obs_part_mean, label='4Obs')
plt.plot(thetas, AUC_6Obs_part_mean, label='6Obs')
plt.vlines(
    175,
    min(np.concatenate([AUC_mjjj_part_mean, AUC_4Obs_part_mean,
                   AUC_6Obs_part_mean])),
    max(np.concatenate([AUC_mjjj_part_mean, AUC_4Obs_part_mean,
                   AUC_6Obs_part_mean])),
    label='Truth')
plt.xlabel("mT (GeV)")
plt.ylabel("AUC")
plt.legend()
# plt.savefig("DCTR_1D_mT-EnsembleDataProcessing-Particle-Level-AUC-120-Run-Average.png")
plt.show()
```
