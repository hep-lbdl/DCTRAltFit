# Load Data

## Utilities

```python
# normalize pT and center (y, phi)
def normalize(x):
    mask = x[:,0] > 0
    yphi_avg = np.average(x[mask,1:3], weights=x[mask,0], axis=0)
    x[mask,1:3] -= yphi_avg
    x[mask,0] /= x[:,0].sum()
```

```python
def preprocess_data(X):
    for x in X:
        normalize(x)
    
    # Remap PIDs to unique values in range [0,1]
    remap_pids(X, pid_i=3)
    return X
```

# Defining reweighting function

$w(x_{T,i},\theta)=((f(x_{T,i},\theta)/(1-f(x_{T,i},\theta)))$

Takes observable from simulation ${\bf \theta_0}$ and weights it to observable from data (target) ${\bf \theta_1}$.

```python
def reweight(x, val): #from NN (DCTR)
    
    #Creating theta_prime
    alphaS = np.ones(shape =x.shape[0:2])*val # Fitting parameter
    aLund = np.ones(shape =x.shape[0:2])*0.68 # Fixed at default
    probStoUD = np.ones(shape =x.shape[0:2])*0.217 # Fixed at default
    theta_prime = np.stack((alphaS, aLund, probStoUD), axis = 2)
    
    data = np.concatenate((x, theta_prime), axis =2)
    
    f = dctr.model.predict(data, batch_size = 1000) # Use dctr.model.predict_on_batch(d) when using outside training
    weights = (f[:,1])/(f[:,0])
    return weights, data
```
